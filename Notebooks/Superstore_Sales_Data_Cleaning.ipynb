# Superstore_Sales_Dataset
## Data Cleaning


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style = 'whitegrid')
```

```python
#Read CSV File
sales = pd.read_csv("E:/Superstore_Sales_Dataset/train.csv")

#Print first 10 records
sales.head(10)
```

### Cleaning column names 

```python
#Clean column names to lowercase and replace white spaces with underscores to make it as single word
sales.columns = sales.columns.str.lower()
sales.columns

sales.columns = sales.columns.str.replace(" ","_").str.replace("-","_")
sales.columns
```

### Cleaning Date columns

```python
#Clean the date columns to datetime format - order_date and ship_date
#coerce = if date is invalid it will replace with NaT like NaN
sales['order_date'] = pd.to_datetime(sales['order_date'], errors = 'coerce', dayfirst=True)
sales['ship_date'] = pd.to_datetime(sales['ship_date'], errors = 'coerce', dayfirst=True)

#Check results
sales[['order_date','ship_date']].head()
```

### Check for missing and duplicate data

```python
#check null values
sales.isnull().sum()
```

```python
#Revert fillna because column had num data type and "unknown" fill made it string
original_postal_code = pd.read_csv("E:/Superstore_Sales_Dataset/train.csv")['Postal Code']

sales['postal_code'] = original_postal_code
```

```python
#Check null values
sales[sales['postal_code'].isnull()]
```

```python
#Converting to str data type and replace null with UNKNOWN
sales['postal_code']=sales['postal_code'].astype(str)
sales['postal_code']=sales['postal_code'].replace('NaN','Unknown')
```

```python
#Check results
sales[sales['postal_code'].isnull()]
```

```python
#Check if there are any duplicate rows
sales.duplicated().sum()
```

```python
#Save to local
sales.to_csv("E:/Superstore_Sales_Dataset/sales_data.csv", index = False)
```
